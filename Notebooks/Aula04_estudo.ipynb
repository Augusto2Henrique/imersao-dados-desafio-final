```python
import pandas as pd

url_dados = 'https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_experimentos.zip?raw=true'
dados = pd.read_csv(url_dados, compression = 'zip')

dados
```

```python
composto = dados['composto'].value_counts()[0:5][::-1] #5 primeiros items em ordem crescente
cod_compostos = composto.index
cod_compostos
```

```python
import pandas as pd #ferramentas de análises
import seaborn as sb #ferramenta de estatística
import matplotlib.pyplot as plt #ferramenta gráfica

url_dados = 'https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_experimentos.zip?raw=true'
tabela_principal = pd.read_csv(url_dados, compression = 'zip')
tabela_principal
```

```python
tabela_principal.head() #pega os 5 primieiros itens da lista
```

```python
muda_nome = {'droga': 'composto'}
tabela_principal.rename(columns = muda_nome, inplace=True)
tabela_principal.head()
```

```python
pd.crosstab(tabela_principal['dose'], tabela_principal['tempo']) #tabela de fraquencia com valor absoluto
```

```python
pd.crosstab([tabela_principal['dose'],
             tabela_principal['tempo']],
            tabela_principal['tratamento'],
            normalize='index') #normalize=True (porcentagem em relação ao total)
                               #normalize='index' (porcentagem em relação a linha)
```

```python
pd.crosstab([tabela_principal['dose'], tabela_principal['tempo']], tabela_principal['tratamento'], values=tabela_principal['g-0'], aggfunc='mean')
```

```python
sb.lmplot(data = tabela_principal,
          x = 'g-0',
          y = 'g-8',
          line_kws = {'color': 'red'},
          col = 'tratamento', #analiza a coluna 'tratamento' e plota ela como coluna
          row = 'tempo') #analiza a coluna 'tempo' e plota ela como uma linha

plt.show()
```

```python
correlacao = tabela_principal.loc[:, 'g-0': 'g-771'].corr()
correlacao
```

*   Correlação próxima de |1| -> variaveis que se influenciam
*   Correlação próxima de |0| -> variaveis que não se influenciam




```python
correlacao = tabela_principal.loc[:, 'g-0': 'g-50'].corr()
import numpy as np

mask = np.triu(np.ones_like(correlacao, dtype=bool))
sb.set()

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(16, 9))

# Generate a custom diverging colormap
cmap = sb.diverging_palette(230,
                            20,
                            as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sb.heatmap(correlacao,
           mask=mask,
           cmap=cmap,
           center=0,
           square=True, 
           linewidths=.5, 
           cbar_kws={"shrink": .5})

plt.show()
```
